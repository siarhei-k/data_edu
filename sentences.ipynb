Найти два предложения, которые ближе всего по смыслу к расположенному в самой первой строке. 

```python
import re
import numpy as np
from scipy.spatial import distance
```

```python
with open('/home/sk/Downloads/sentences.txt') as st:
    original = st.read().splitlines()
    tokenized = []
    for line in range(len(original)):
        original[line] = re.split('[^a-z]', original[line].lower())
        original[line] = list(filter(None, original[line]))
        for word in original[line]:
            tokenized.append(word)
            
tokenized_unique = list(set(tokenized))
tokenized_unique.sort()
```

```python
original
```

```python
n = len(original) # n — число предложений
d = len(tokenized_unique) # d — число различных слов в предложениях
```

```python
d_index = dict()
for i in range(d):
    d_index[i] = tokenized_unique[i]
```

```python
d_index
```

```python
matrix = np.ones((n, d))

for i in range(n):
    for j in range(d):
        matrix[i, j] = original[i].count(d_index[j])
```

```python
result = []

for row in range(1, n):
    cousine = distance.cosine(matrix[0, :], matrix[row, :])
    result.append(cousine)
```

```python
result = []
d_result = dict()

for row in range(1, n):
    cousine = distance.cosine(matrix[0, :], matrix[row, :])
    result.append(cousine)
    d_result[cousine] = row
```

```python
d_result
```

```python
values = list(d_result.keys())
values.sort()

answer = [d_result[values[0]], d_result[values[1]]]
answer.sort()
```

```python
submission = open('submission-1.txt', 'w')
submission.write(str(answer[0]) + ' ' + str(answer[1]))
submission.close()
```
